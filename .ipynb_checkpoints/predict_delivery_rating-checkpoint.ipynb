```python
to_predict = [
    'Delivery',
    'Distance',
    'Items in order',
    'Bill subtotal',
    'Total',
    'KPT duration (minutes)',
    'Rider wait time (minutes)',
    'Subzone',
    'City'
]
```

```python
plt.figure(figsize=(6,3))
sns.histplot(df['KPT duration (minutes)'], kde=True)
plt.title('Distribution of KPT Duration (minutes)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Delivery', hue='high_rating')
plt.title('Delivery Type Counts by High Rating')
plt.show()
```

```python
df[to_predict].head()
```

```python
df['high_rating'] = df['Rating'].apply(lambda x: 1 if x >= 4 else 0)
```

```python
df['high_rating']
```

```python
df = df.dropna(subset=['Rating'])
```

```python
df['KPT duration (minutes)'].fillna(df['KPT duration (minutes)'].median(), inplace=True)
df['Rider wait time (minutes)'].fillna(df['Rider wait time (minutes)'].median(), inplace=True)
```

```python
df[to_predict].isna().sum()
```

```python
df['high_rating'].value_counts()
```

```python
x = df[to_predict] 
y = df['high_rating'] 
```

```python
x.head()
```

```python
y.value_counts()
```

```python
df.loc[:, 'Items in order'] = df['Items in order'].apply(lambda x: len(str(x).split(','))) 
```

```python
df.loc[:, 'Distance'] = df['Distance'].str.replace('<', '')
df.loc[:, 'Distance'] = df['Distance'].str.replace('km', '')
```

```python
df['Items in order'].head()
```

```python
x_encoded = pd.get_dummies(df[to_predict], columns=['Delivery', 'City', 'Subzone'], drop_first=True)
```

```python
x_train,x_test, y_train, y_test = train_test_split(x_encoded,y,test_size=0.3,random_state=67) 
```

```python
model = LogisticRegression(max_iter=1000, class_weight='balanced')
model.fit(x_train,y_train) 
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
f1 = f1_score(y_test, y_pred)
print("F1 Score:", f1)
```
